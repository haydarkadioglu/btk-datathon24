```python
import pandas as pd
import numpy as np
import random
```

```python
data = pd.read_csv("datathon-2024/clean-test-data-v1.csv")
org = pd.read_csv('datathon-2024/test_x.csv')
```

küçük büyük harf uyumsuzluğu giderme


```python
# Metin tabanlı sütunları seçip, büyük harfe çevir
for column in data.columns:
    if data[column].dtype == 'object':  # Eğer sütun tipi 'object' ise, bu bir metin sütunudur
        data[column] = data[column].str.upper()  # Büyük harfe çevir

data.replace("-", 0)
data.fillna(0)
```

# Yaş kolonunu eklemee


```python
turkce_aylar = ['Ocak', 'Subat', 'Mart', 'Nisan', 'Mayis', 'Haziran', 'Temmuz', 'Agustos', 'Eylul', 'Ekim', 'Kasim', 'Aralik']
ingilizce_aylar = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
```

```python
data['Dogum Tarihi'] = org['Dogum Tarihi']
```

```python
for tr_ay, en_ay in zip(turkce_aylar, ingilizce_aylar):
    data['Dogum Tarihi'] = data['Dogum Tarihi'].str.replace(tr_ay, en_ay, regex=True)

data['Dogum Tarihi'] = pd.to_datetime(data['Dogum Tarihi'], errors='coerce')
```

```python
# data['Basvuru Yili'] = pd.to_numeric(data['Basvuru Yili'], errors='coerce')


data['Yas'] = data['Basvuru Yili'] - data['Dogum Tarihi'].dt.year
```

```python
data.to_csv('datathon-2024/clean-test-data-yas.csv')
```
